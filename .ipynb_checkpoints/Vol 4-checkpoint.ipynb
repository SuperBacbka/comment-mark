```python
stopwords_ru = set([
    'а', 'без', 'был', 'была', 'были', 'быть', 'в', 'во', 'всё', 'все', 'да', 'для', 'до',
    'его', 'ее', 'если', 'есть', 'же', 'за', 'и', 'из', 'или', 'их', 'к', 'как', 'ко', 'когда',
    'который', 'ли', 'либо', 'мне', 'может', 'мы', 'на', 'надо', 'наш', 'не', 'него', 'нее',
    'нет', 'ни', 'но', 'ну', 'о', 'об', 'однако', 'он', 'она', 'они', 'оно', 'от', 'очень',
    'по', 'под', 'при', 'с', 'со', 'так', 'также', 'там', 'те', 'тем', 'то', 'того', 'тоже',
    'той', 'только', 'том', 'ты', 'у', 'уже', 'хотя', 'чего', 'чей', 'чем', 'что', 'чтобы',
    'чье', 'эта', 'эти', 'это', 'я'
])
```

```python
df['comment'] = df['comment'].apply(
    lambda text: ' '.join([
        word for word in str(text).lower().split()
        if word not in stopwords_ru and word.isalpha() and len(word) >= 3
    ])
)
```

```python

category_phrases = {
    'Нравится скорость отработки заявок': ['быстро', 'оперативно', 'незамедлительно', 'мгновенно'],
    'Нравится качество выполнения заявки': ['качественно', 'отлично сделано', 'хорошо выполнено', 'без нареканий'],
    'Нравится качество работы сотрудников': ['вежливый', 'профессионал', 'компетентен', 'отзывчивый', 'специалист'],
    'Понравилось выполнение заявки': ['заявка выполнена', 'сделали', 'устранили', 'решили проблему'],
    'Вопрос решен': ['решено', 'вопрос закрыт', 'разобрались', 'проблема решена', 'устранили']
}

```

```python
for category, phrases in category_phrases.items():
    df[category] = df['comment'].apply(lambda text: int(any(p in text for p in phrases)))
```

```python
df[category].unique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

labal = {}

for category in category_phrases:
    subset = df[df[category] == 1]
    if not subset.empty:
        vec = TfidfVectorizer(ngram_range=(1, 3), max_features=1000)
        X = vec.fit_transform(subset['comment'])
        tfidf = X.sum(axis=0).A1
        features = vec.get_feature_names_out()
        top = sorted(zip(features, tfidf), key=lambda x: x[1], reverse=True)[:20]
        labal[category] = [phrase for phrase, _ in top]
        labels, scores = zip(*top)
        plt.figure(figsize=(20, 10))
        sns.barplot(x=list(scores), y=list(labels))
        plt.title(f"TF-IDF фразы для категории: {category}")
        plt.xlabel("TF-IDF вес")
        plt.ylabel("Фраза")
        plt.tight_layout()
        plt.show()


    
```

```python
for cat, phrases in labal.items():
    print(f"\n{cat}:")
    for p in phrases:
        print(f"  - {p}")
```

```python
for cat, word in labal.items():
    labal[cat] = [w.replace('спасибо', '') for w in word]
```

```python
for cat in labal:
    labal[cat] = [w for w in labal[cat] if w]
```

```python
labal_clean = {}
for cat, word in labal.items():
   
    cleaned = [w.strip() for w in word if w.strip()]
    labal_clean[cat] = list(dict.fromkeys(cleaned))
```

```python
from collections import defaultdict

phrase_map = defaultdict(list)

for category, phrases in labal_clean.items():
    for phrase in phrases:
        phrase_map[phrase.strip()].append(category)

cross_phrases = {p: cats for p, cats in phrase_map.items() if len(cats) > 1}

for phrase, cats in cross_phrases.items():
   
    print(f"«{phrase}» встречается в категориях:  {', '.join(cats)}" , end = "\n" )
    print(100 * '-')
```

```python
# Разметка по словарю
for category in labal_clean:
    df[category] = 0
    df[f'{category}_matched'] = ''

for idx, row in df.iterrows():
    text = str(row['comment']).lower()
    for category, phrases in labal_clean.items():
        matched = [phrase for phrase in phrases if phrase in text]
        if matched:
            df.at[idx, category] = 1
            df.at[idx, f'{category}_matched'] = ', '.join(matched)
```

```python
from sklearn.model_selection import train_test_split

X = df['comment']
y = df[label_columns]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python

tfidf = TfidfVectorizer(max_features=1000, ngram_range=(1, 2))
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

# Обучение модели
model = OneVsRestClassifier(LogisticRegression(max_iter=300, class_weight='balanced'))
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_pred_proba = model.predict_proba(X_test_tfidf)

roc_auc_scores = {}
for i, label in enumerate(label_columns):
    if len(set(y_test[label])) > 1:
        roc_auc_scores[label] = roc_auc_score(y_test[label], y_pred_proba[:, i])

mean_roc_auc = sum(roc_auc_scores.values()) / len(roc_auc_scores)

print("ROC-AUC по категориям:", roc_auc_scores)
print("Средний ROC-AUC:", round(mean_roc_auc, 4))
```

```python

y_train_pred_proba = model.predict_proba(X_train_tfidf)

roc_auc_train = {}
for i, label in enumerate(label_columns):
    if len(set(y_train[label])) > 1:
        roc_auc_train[label] = roc_auc_score(y_train[label], y_train_pred_proba[:, i])

mean_train_roc_auc = sum(roc_auc_train.values()) / len(roc_auc_train)

print("ROC-AUC на обучении:", mean_train_roc_auc)
print("Разница:", round(mean_train_roc_auc - mean_roc_auc, 4))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

pipeline = make_pipeline(
    TfidfVectorizer(max_features=3000, ngram_range=(1, 2)),
    OneVsRestClassifier(LogisticRegression(max_iter=1500, class_weight='balanced'))
)


scores = cross_val_score(pipeline, X, y, cv=5, scoring='roc_auc_ovr_weighted')

print("ROC-AUC по фолдам:", scores)
print("Среднее значение:", scores.mean())
```
